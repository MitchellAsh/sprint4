Import necessary tools and the dataframe

```python
import pandas as pd
import plotly.express as px
vehicles_path = '../vehicles_us_data.csv'
df = pd.read_csv(vehicles_path, keep_default_na=False)
```

Print out basic information about the data

```python
print(df.head())
print()
df.info()
print()
df.isnull().sum()
print()
df.describe()
```

Check for duplicates and drop them

```python
df.duplicated().sum()
df.drop_duplicates()
```

Plot histograms and scatterplots for visualization

```python
fuel_counts = df['fuel'].value_counts().reset_index()

fuel_counts.columns = ['fuel', 'count']

fig = px.bar(fuel_counts, x='fuel', y='count', title='Car Count by Fuel Type')
fig.show()
```

```python
fig = px.histogram(df, x='days_listed', nbins=30, title='Days Listed Distribution')
fig.show()
```

```python
fig = px.box(df, x='condition', y='price', title='Condition vs Price')
fig.show()
```

```python
fig = px.scatter(df, x='odometer', y='price', title='Odometer vs Price')
fig.update_layout(
    xaxis=dict(
        type='log',  # Set the x-axis to a logarithmic scale
    )
)
fig.show()
```
